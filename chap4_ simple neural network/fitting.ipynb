```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# 
x_train = np.random.uniform(-2 * np.pi, 2 * np.pi, size=(5000, 1))
y_train = np.sin(x_train) + np.random.normal(0, 0.1, size=(5000, 1))

x_test = np.linspace(-2 * np.pi, 2 * np.pi, 1000).reshape(-1, 1)
y_test = np.sin(x_test)

# 
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse')

# 
history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_split=0.2)

# 
loss = model.evaluate(x_test, y_test)
print("Test Loss:", loss)

# 
plt.figure(figsize=(8, 6))
plt.scatter(x_test, y_test, color='blue', label='True Function')
plt.scatter(x_test, model.predict(x_test), color='red', label='Predicted Function')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitting Sin Function')
plt.show()
```
